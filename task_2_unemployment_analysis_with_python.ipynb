```python
import pandas as pd
df=pd.read_csv("Unemployment in India.csv")
print(df.head())
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df.dropna())
```

```python
df.columns=df.columns.str.strip()
print(df.columns)
```

```python
print(df.columns.tolist())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.histplot(df["Estimated Unemployment Rate (%)"],kde= True)
plt.title("Distribution of Estimated Unemployment Rate")
plt.xlabel("Estimated Unemployment Rate(%)")
plt.ylabel("count")
plt.show()
```

```python
(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,6))
sns.barplot(
    x='Region',
    y='Estimated Unemployment Rate (%)',
    data=df

)
plt.xticks(rotation=90)
plt.title("Estimated Unemployment Rate by Region")
plt.xlabel("Region")
plt.ylabel("Estimated Unemployment Rate(%)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,6))
sns.barplot(
    x='Region',
    y='_Estimated_Labour_participation_Rate_(%)',
    data=df
)
plt.xticks(rotation=90)
plt.title("Estimated Labour Participation Rate by Region")
plt.xlabel("Region")
plt.ylabel("Estimated Labour Participation Rate(%)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
numeric_df=df.select_dtypes(include=[np.number])
corr_matrix=numeric_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True, cmap='coolwarm',linewidths=0.5
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Estimated Labour Participation Rate (%)',
    y='Estimated Unemployment Rate (%)',
    data=df
)
plt.title("Estimated Participation Rate vs Unemployment Rate")
plt.xlabel("Estimated Labour Participation Rate(%)")
plt.ylabel("Estimated Unemployment Rate(%)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.boxplot(
    x='Area',
    y='Estimated Unemployment Rate (%)',
    data=df
)
plt.title("Urban vs Rural Estimated Unemployment Rate")
plt.xlabel("Area")
plt.ylabel("Estimated Unemployment Rate(%")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(
    x='Region',
    y='Estimated Unemployment Rate (%)',
    data=df
)
plt.xticks(rotation=90)
plt.title("State wise Unemployment Rate")
plt.xlabel("Region")
plt.ylabel("Estimated Unemployment Rate(%)")
plt.show()
```
